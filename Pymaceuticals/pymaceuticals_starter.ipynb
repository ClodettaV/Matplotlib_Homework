## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

mouse_metadata

# Display the mouse metadata table for preview
```

```python
# Display the study resuts table for preview
study_results
```

```python
# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata, study_results, on = 'Mouse ID', how = 'outer')
combined_data
```

```python
# Checking the number of mice.

number_of_mice = combined_data['Mouse ID'].str.len()
number_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

mouse_ID_duplicates = combined_data[combined_data.duplicated(subset= ['Mouse ID', 'Timepoint'], keep = False)]
mouse_ID_duplicates = mouse_ID_duplicates[['Mouse ID', 'Timepoint']]
mouse_ID_duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = combined_data.drop_duplicates(subset = ['Mouse ID', 'Timepoint'])
clean_data
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
# display dataframe with the Drug Regimen column filtered by "Capomulin" 
Capomulin_list = clean_data[clean_data['Drug Regimen']=="Capomulin"]
Capomulin_list
```

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
x_axis = clean_data.loc[clean_data["Mouse ID"] == "s185"]
x_axis = x_axis["Timepoint"]
y_axis = clean_data.loc[clean_data["Mouse ID"] == "s185"]
y_axis = y_axis["Tumor Volume (mm3)"]

plt.plot(x_axis, y_axis)
plt.title("Tumor Volume with Capomulin Treatment of Mouse s185")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

mouse_weight = Capomulin_list["Weight (g)"]
tumor_volume = Capomulin_list["Tumor Volume (mm3)"]
plt.scatter(mouse_weight, tumor_volume)
plt.title("Average Mouse Weight versus Average Tumor Volume for Capomulin Regimen")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr_coeff = st.pearsonr(mouse_weight, tumor_volume)
print(f"The correlation between both weight and tumor volume is {round(corr_coeff[0],2)}")

slope, int, r, p, std_err = st.linregress(mouse_weight, tumor_volume)
fit = slope * mouse_weight + int
plt.scatter(mouse_weight, tumor_volume)
plt.plot(mouse_weight, fit, "--", color="r")
plt.title("Average Mouse Weight versus Average Tumor Volume for Capomulin Regimen")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
```
